```python
value_counts = df['Officer_Name'].value_counts()

# Display the unique values and their counts
print(value_counts)
```

```python
#Group by 'Officer_Name' and calculate relevant statistics
employee_performance = df.groupby('Officer_Name').agg(
    Total_Activities=('Transaction_ID', 'count'),
    Avg_Activity_Duration=('Activity_Duration', 'mean'),
    Avg_Pollution_Level=('Pollution_Level_Detected', 'mean'),
    Avg_Community_Feedback=('Community_Feedback_Rating', 'mean'),
    Compliance_Rate=('compliance_status', lambda x: (x == 'Compliant').mean())
)
employee_performance
```

```python
# Normalizing the data to put them on the same scale
employee_performance['Normalized_Activities'] = employee_performance['Total_Activities'] / employee_performance['Total_Activities'].max()
employee_performance['Normalized_Duration'] = employee_performance['Avg_Activity_Duration'] / employee_performance['Avg_Activity_Duration'].max()
employee_performance['Normalized_Pollution'] = employee_performance['Avg_Pollution_Level'] / employee_performance['Avg_Pollution_Level'].max()
employee_performance['Normalized_Compliance'] = employee_performance['Compliance_Rate'] / employee_performance['Compliance_Rate'].max()
```

```python
employee_performance
```

```python
# Assuming that higher pollution detection and lower activity duration are better
# Adjust the score formula depending on the importance of each factor
# You may want to give different weights (e.g., 0.4 for compliance, 0.3 for pollution level, etc.)
employee_performance['Performance_Score'] = (
    0.4 * employee_performance['Normalized_Activities'] +
    0.3 * employee_performance['Normalized_Compliance'] +
    0.2 * (1 - employee_performance['Normalized_Duration']) +  # Assuming lower duration is better
    0.1 * employee_performance['Normalized_Pollution']
)
```

```python
employee_performance
```

```python
# Sort by the performance score to find the best employee
best_employee = employee_performance.sort_values(by='Performance_Score', ascending=False).head(1)

print("Best Employee based on adjusted performance score:")
print(best_employee)
```
